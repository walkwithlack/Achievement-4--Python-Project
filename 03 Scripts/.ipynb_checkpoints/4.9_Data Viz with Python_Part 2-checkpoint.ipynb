# 2. Creating bar chart, histogram, scatterplot and line chart using a sample subset

```python
import matplotlib.pyplot as plt
from matplotlib import cm

# Count orders by day of the week
order_counts = df['orders_day_of_week'].value_counts().sort_index()

# Use Set2 colormap with 7 colors
colors = cm.get_cmap('Set2', 7).colors

# Plotting
fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(order_counts.index, order_counts.values, color=colors)

# Axis labels and title
ax.set_xlabel('Day of the Week', fontsize=12)
ax.set_ylabel('Number of Orders', fontsize=12)
ax.set_title('Orders by Day of the Week', fontsize=14)

# Create weekday labels (0 = Sunday, 6 = Saturday)
day_labels = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
ax.set_xticks(order_counts.index)
ax.set_xticklabels([day_labels[int(i)] for i in order_counts.index], rotation=45)

# Exporting bar chart
fig.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'bar_orders_dow_green_yellow.png'))

plt.tight_layout()
plt.show()
```

```python
# Create the histogram for prices
plt.figure(figsize=(8, 5))
hist = df['prices'].plot.hist(bins=30, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Price', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Prices', fontsize=14)

# Export viz
hist.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_prices_bin_30.png'))

# Improve layout
plt.tight_layout()
plt.show()
```

- Creating line by first creating a sample subset

```python
#setting seed and creating random list of floats between 0 and 1 holding True/False values to the test np.random.rand()<=7
np.random.seed(4)
dev = np.random.rand(len(df)) <= 0.7
```

```python
# use list to split dataframe into two samples. 
#storing 70% of the sample in the dataframe big; 
#storing 30% of the sample in the dataframe small
big = df[dev]
small = df[~dev]
```

```python
#checking len of large df and subsets we created

len(df)
```

```python
len(big) + len (small)
```

```python
# reducing samples to necessary columns to speed up processing time
df_2 = small[['orders_day_of_week','prices']]
```

```python
day_labels_ordered = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

# Map numeric days to weekday names
df['orders_day_of_week_name'] = df['orders_day_of_week'].map(lambda x: day_labels_ordered[int(x)])

# Convert to ordered categorical type with correct weekday order
df['orders_day_of_week_name'] = pd.Categorical(
    df['orders_day_of_week_name'],
    categories=day_labels_ordered,
    ordered=True
)

plt.figure(figsize=(10, 6))

# Create the line chart with ordered categorical x-axis
line = sns.lineplot(data=df, x='orders_day_of_week_name', y='prices', marker='o')

# Label axes and title
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Average Price', fontsize=12)
plt.title('Average Prices by Day of the Week', fontsize=14)
plt.xticks(rotation=45)

plt.tight_layout()

# Exporting the figure
line.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'line_prices_dow.png'))

plt.show()
```

# 3. Order_hour_of_day and sales histogram

```python
#creating histogram
plt.figure(figsize=(10, 6))
hist_order_hour = df['order_hour_of_day'].plot.hist(
    bins=24,  # 0â€“23 hours = 24 bins for clarity
    color='skyblue',
    edgecolor='black'
)

# Add labels and title
plt.xlabel('Hour of Day', fontsize=12)
plt.ylabel('Number of Orders', fontsize=12)
plt.title('Order Distribution by Hour of Day', fontsize=14)

# Add vertical lines between hours
plt.xticks(range(0, 24, 1))  # one tick per hour
plt.grid(axis='x', which='major', linestyle='--', alpha=0.7)

# exporting histogram
hist_order_hour.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_order_hour.png'))

# tighten layout
plt.tight_layout()
plt.show()
```

Since we have in our dataframe one row per item ordered, this histogram shows the distribution of all orders across 24 hours. We see the peak in orders between 10 AM and 4 PM.

# 4. Distribution of Orders among Loyalty flags

```python
#creating bar chart
proportions = df['loyalty_flag'].value_counts(normalize=True)
bar_loyalty_order = proportions.plot.bar(
    color=['skyblue', 'gray', 'darkgray'], figsize=(8,5)
)
# add labels
plt.xlabel('Loyalty Flag')
plt.ylabel('Proportion of Orders')
plt.title('Proportion of Orders by Loyalty Flag')
plt.xticks(rotation=0)
plt.tight_layout()

# exporting bar chart
bar_loyalty_order.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_loyalty_order.png'))

plt.show()
```

Most customers are regular customers, followed by loyal customers, then new customers

# 5. Prices and Hour of day relationship

```python
# Set seed and create random boolean mask for ~80% sample
np.random.seed(4)
bm = np.random.rand(len(df)) <= 0.8
```

```python
# Split the dataframe into two samples
sample_big = df[bm]      
sample_small = df[~bm]
```

```python
# check lengths
print(len(df), len(sample_big) + len(sample_small))
```

```python
# Reduce to needed columns to speed up processing
df_reduced = sample_small[['order_hour_of_day', 'prices']]
```

```python
# Create line chart for avg prices by hour

plt.figure(figsize=(10, 6))
linechart = sns.lineplot(data=df_reduced, x='order_hour_of_day', y='prices')

plt.xlabel('Order Hour of Day', fontsize=12)
plt.ylabel('Average Price', fontsize=12)
plt.title('Average Prices by Hour of Day', fontsize=14)
plt.tight_layout()

# exporting linechart
linechart.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_price_hour.png'))

plt.show()
```

Highest prices at night, lowest in the morning, recovery in the afternoon

# 6. Age and Family situation relationship

```python
# Set seed and create random boolean mask for ~80% sample
np.random.seed(4)
mask = np.random.rand(len(df)) <= 0.8
```

```python
# Split into samples
sample_big = df[mask]      
sample_small = df[~mask]
```

```python
# Check lengths 
print(len(df), len(sample_big) + len(sample_small))
```

```python
# Reduce to necessary columns
demo_df = sample_small[['age', 'household_size']]
```

```python
# Group by age and calculate average household size
age_family = demo_df.groupby('age')['household_size'].mean().reset_index()
```

```python
# create line chart
age_fam_line = sns.lineplot(data=age_family, x='age', y='household_size')
```

Lows in the 20s and 80 (most people being single / divorced or widowed), interestingly at 60 too (too young for most people to die. divorce-era?) where also have peak (families where children are still living with parents - economic reasons or parents became parents later in life). 

```python
# exporting linechart
age_fam_line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'age_fam_line.png'))
```

# 7. Age and Income relationship

```python
# Set random seed and sampling mask (~70/30 split)
np.random.seed(4)
mask = np.random.rand(len(df)) <= 0.7
```

```python
# Create sample
sample_big = df[mask]
sample_small = df[~mask]
```

```python
# Reduce to necessary columns
scatter_df = sample_small[['age', 'income']]
```

```python
# Create scatterplot
age_income = sns.scatterplot(data=scatter_df, x='age', y='income')
```

Outliers at all ages (we saw this already in Part 1 of our Task) but especially after 40 (where career differences and inheritances really kick in!). Some people migrate to some higher income in their 40s (what is called "social mobility"). For most people - at least at IC :)) - however, income remains the same throughout their life.

```python
# exporting scatterplot

file_path = os.path.join(path, '04 Analysis', 'Visualizations', 'age_income_plot.png')
plt.savefig(file_path, dpi=300, bbox_inches='tight')
```
